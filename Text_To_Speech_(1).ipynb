# Text To Speech
## This notebook outlines the concepts invovled in Converting a piece of text into speech

Google Text to Speech API - **gTTS**

### Features

- Customizable speech-specific sentence tokenizer that allows for unlimited lengths of text to be read, all while keeping proper intonation, abbreviations, decimals

- Customizable text pre-processors which can, for example, provide pronunciation corrections

- Automatic retrieval of supported languages

### Steps
- 1. Install the library
- 2. Import the library
- 3. Create Text-to-Speech object
- 4. Create a text
- 5. Send the text for conversion

### Import the library

```python
from gtts import gTTS
import os
import IPython.display as ipd
```

### Sample text to convert

```python
# This is the text that will be converted to speech
text = "Hello! Welcome to the Text to Speech conversion tutorial. \
        This is powered by Google Text to Speech API."
```

### Specify language

```python
language = 'en'
slow = False
```

### Convert Text to Speech

```python
tts = gTTS(
    text=text,
    lang=language,
    slow=slow,
    tld='com'
)

print("Converted Text to Speech", tts)
```

### Save the converted speech

```python
output_file = "output_mp3.mp3"
tts.save(output_file)
```

### Play the speech

```python
print('Playing Audio...')
ipd.Audio(output_file)
```

### Note on Turkmen Language Support
gTTS does not have a dedicated Turkmen language code.
Turkish ('tr') is used as the closest phonetic alternative,
as both languages share Turkic roots and similar pronunciation patterns.
This is a common workaround used in production TTS applications.

```python
from gtts.lang import tts_langs

# Search for Turkmen or Turkish
langs = tts_langs()
for code, name in langs.items():
    if "turk" in name.lower() or "tk" in code.lower():
        print(f"  '{code}' → {name}")
```
